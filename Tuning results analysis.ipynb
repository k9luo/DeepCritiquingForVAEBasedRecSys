```python
import ast
import os
import pandas as pd
pd.set_option('display.max_columns', 50)
```

```python
def load_tuning_results(folder_path):
    csv_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f)) and f.endswith('.csv')]
    
    df = pd.DataFrame()
    for record in csv_files:
        df = df.append(pd.read_csv(record))
    return df
```

```python
df_beer_rating = load_tuning_results(folder_path='tables/beer_rating_tuning/')
df_beer_rating = df_beer_rating.loc[df_beer_rating['lambda_keyphrase'] == df_beer_rating['lambda_latent']]
metric = 'NDCG'
df_beer_rating[metric+'_Score'] = df_beer_rating[metric].map(lambda x: ast.literal_eval(x)[0])
df_beer_rating = df_beer_rating.sort_values(by=metric+'_Score', ascending=False)
```

# Find Best Intermediate Hyperparameters for Beer

### 1st Selection Approach

```python
df_beer_rating['rating_rank'] = range(len(df_beer_rating))
df_beer_explanation['explanation_rank'] = range(len(df_beer_explanation))
```

```python
df_beer = pd.merge(df_beer_rating, df_beer_explanation,  how='inner', on=['rank', 'beta', 'lambda_l2', 
                                                                          'lambda_keyphrase', 'lambda_latent', 
                                                                          'lambda_rating', 'learning_rate', 
                                                                          'epoch', 'corruption'])
```

```python
df_beer
```

```python
df_beer['total_rank'] = df_beer['rating_rank'] + df_beer['explanation_rank']
```

```python
df_beer.sort_values(by='total_rank', ascending=True)
```

### 2nd Selection Approach

```python
df_beer_rating['normalize'] = df_beer_rating['NDCG_Score'] / df_beer_rating['NDCG_Score'].max()
df_beer_explanation['normalize'] = df_beer_explanation['NDCG_Score'] / df_beer_explanation['NDCG_Score'].max()
df_beer = pd.merge(df_beer_rating, df_beer_explanation,  how='inner', on=['rank', 'beta', 'lambda_l2', 
                                                                          'lambda_keyphrase', 'lambda_latent', 
                                                                          'lambda_rating', 'learning_rate', 
                                                                          'epoch', 'corruption'])
df_beer['normalize'] = df_beer['normalize_x'] * df_beer['normalize_y']
df_beer.sort_values(by='normalize', ascending=False)
```
